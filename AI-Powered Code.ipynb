```python
# Manual implementation of the sorting function.

def sort_by_age(data):
    """
    This function sorts a list of dictionaries by the 'age' key.
    We're using a lambda to tell Python how to do it.
    """

    # The lambda function just fetches the value of the 'age' key for each item.
    return sorted(data, key=lambda x: x['age'])

# --- My dictionary list ---
my_data = [
    {'name': 'Yonela', 'age': 15, 'city': 'Gqeberha'},
    {'name': 'Amany', 'age': 24, 'city': 'Nairobi'},
    {'name': 'Kim', 'age': 37, 'city': 'Paris'},
    {'name': 'David', 'age': 58, 'city': 'Berlin'},
    {'name': 'Lefa', 'age': 23, 'city': 'Pretoria'},
    {'name': 'Cindy', 'age': 41, 'city': 'Tokyo'}
]

print("The original List:")
print(my_data)

# Now to sort data in my dictionary according to age
sort_people = sort_by_age(my_data)

print("\nSorted List by 'age':")
print(sort_people)
```
